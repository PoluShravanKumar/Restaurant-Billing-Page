```python
## Restaurant Menu with different types of food items
menu ={     
     "🥣🥣SOUP ITEMS🥣🥣:": {
        "SoupIndo Chinese Manchow soup":159,
        "Hot & sour soup":169,
        "Sweetcorn soup":189,
        "La cream de tomato":199,
        "Clear soup":209
    },
        
    "🍗STARTERS🍗":{
        "Subj Harabhara kebab":299,
        "potato and peas tikki":199,
        "Sunehari kebab":319,
        "Masala popad":259        
    },
        
    "🍖INDIAN_NON_VEG_ITEMS🍖" :{
        "Bhatti de murgh tikka":249,                            
        "Murgh pahadi tikka":199,
        "Murgh Malai tikka":259,                                  
        "Murghu tandoori":259,                                  
        "Prawns Koliwada":299
        
    },
        
    "🥬🥬INDUAN_VEG_ITEMS🥬🥬":{
        "Diwani Hundi":149 ,                                
        "Veg Tawa Masala":169,                               
        "Paneer Makhanawala":159,                                
        "Dingari Mashroom Masala":259,                                
        "Dal Thadka":199
    },
        
    "🍜🍜CHINESE_ITEMS🍜🍜":{
        "Kuh-chai Hakka Chilly":159,                            
        "Holysmoke BBQ":249,                             
        "Manchurian dry":249,                            
        "Nachose Chicos":249,                             
        "French fries":229
        
    },
        
    "🍷🍷BEVERAGE_ITEMS🍷🍷":{ 
        "Thumsup 300ml": 70,
        "Sprite 300ml":  70,
        "Coke 300ml" : 70,
        "Redbull 250ml": 125,
        "Monster 250ml":  125
    },
    
    "🍨🍨DESSERT_ITEMS🍨🍨":{
        "Buttur scotch Ice cream 200ml":199,
        "Vanila Ice cream 200ml":149,
        "Gulab Jamun":100,
        "Double ka meeta":100,
        "Chocolate Cake":149
    
    }
}

def display_menu():
    print("+------------------------------------+")
    print("|⛩️ WELCOME TO THE D&S RESTAURANT ⛩️|")
    print("+------------------------------------+")
    print("MENU CATEGORIES :")
    for idx, category in enumerate(menu.keys(), 1):
        print(f"{idx}. {category}")

def get_category_by_index(index):
    categories = list(menu.keys())
    if 1 <= index <= len(categories):
        return categories[index - 1]
    else:
        return None

def place_order():
    order = {}
    total_price = 0

    while True:
        display_menu()
        print("\nOPTIONS:")
        print("1. Order Items" )
        print("2. View your order Bill")
        print("3. Cancel Order")
        print("4. To Pay the Bill ")

        choice = input("Enter your choice from Options: ")

        if choice == '1':
            category_choice = int(input("Enter the number of the category To Order the Items: "))
            selected_category = get_category_by_index(category_choice)
            if selected_category is None:
                print("Invalid category number! Please enter a valid category number.")
                continue
            print(f"\n{selected_category}:")
            for idx, (item, price) in enumerate(menu[selected_category].items(), 1):
                print(f"{idx}. {item} - ₹{price:.2f}")
            item_choice = int(input("Enter the number of the item you want to order (0 to go back to the main menu): "))
            if item_choice == 0:
                continue
            items = list(menu[selected_category].keys())
            if 1 <= item_choice <= len(items):
                selected_item = items[item_choice - 1]
                quantity = int(input("Enter the quantity: "))
                if quantity <= 0:
                    print("Quantity should be greater than 0.")
                    continue
                if selected_item in order:
                    order[selected_item]["quantity"] += quantity
                else:
                    order[selected_item] = {"quantity": quantity, "price": menu[selected_category][selected_item]}
                total_price += menu[selected_category][selected_item] * quantity
                print(f"\n{quantity} {selected_item}(s) added to your order.")
            else:
                print("Invalid item number! Please enter a valid item number.")

        elif choice == '2':
            if not order:
                print("No items in the order yet, Please make order again 😊.")
                continue
            print("\n ************************ YOUR ORDER ***************************")
            print("+----------------------------------------------------------------+")
            print("|{:<35}|{:<15}|{:<10}  |".format("Item", " Quantity", " ₹ Price  "))
            print("+----------------------------------------------------------------+")
            for item, details in order.items():
                
                print("|{:<35}|{} x ₹ {:<10} ₹ {:<9}|".format(item, details['quantity'],details["price"], details['price'] * details['quantity']))
                
            print("+----------------------------------------------------------------+")
            print("|{:<51}|{:<10}  |".format("Total Bill", f" ₹ {total_price}"))
            print("+----------------------------------------------------------------+")

        elif choice == '3':
            confirm = input("Are you sure you want to cancel your order? (yes/no): ")
            if confirm.lower() == 'yes':
                print("Order canceled. ):")
                order = {}
                total_price = 0
            else:
                print("Continuing with the order.")

        elif choice == '4':
            if not order:
                print("No items in the order yet.Please make order again 😊")
                continue
            print("\n ************************ YOUR ORDER ***************************")
            print("+----------------------------------------------------------------+")
            print("|{:<35}|{:<15}|{:<10}  |".format("Item", " Quantity", " ₹ Price  "))
            print("+----------------------------------------------------------------+")
            for item, details in order.items():
                
                print("|{:<35}|{} x ₹ {:<10} ₹ {:<9}|".format(item, details['quantity'],details["price"], details['price'] * details['quantity']))
                
            print("+----------------------------------------------------------------+")
            print("|{:<51}|{:<10}  |".format("Total Bill", f" ₹ {total_price}"))
            print("+----------------------------------------------------------------+")
            tip = float(input("Enter the tip amount (: ,(₹): "))
            total_price += tip
            print(f"Total (including tip): ₹{total_price:.2f}")
            print("Thank you for your order! (:")
            break

        else:
            print("Invalid choice! Please enter a valid option.")

    print("💐💐Thank you for visiting our D&S Restaurant!💐💐")
    print("*"*25,"VISIT AGAIN","*"*25)

if __name__ == "__main__":
    place_order()
```
